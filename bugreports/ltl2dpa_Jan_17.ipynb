# ltl2dpa(ldba) --version from Jan 17th 2018

```python
import spot
spot.setup(show_default='.a')
```

## '(p0 & (p2 R X!p0)) | (Gp1 & (p2 U Xp0))'

```python
f = '(p0 & (p2 R X!p0)) | (Gp1 & (p2 U Xp0))'
```

```python
p = spot.automaton('../Rab4/bin/ltl2dpa --mode=ldba "'+f+'" |')
n = spot.automaton('../Rab4/bin/ltl2dpa --mode=ldba "!({})" |'.format(f))
```

```python
spot.formula(f)
```

The word $\{p1,p2\}^\omega$ is not in $L(f)$ and thus should be rejected by $p$ and accepted by $n$. However, it is rejected by both. So $n$ is incorrect.

```python
p
```

```python
n
```
