```python
f = '(p0) & (G(F(((p0) | (p1)) & ((G(p0)) | (F(G(!(p0))))))))'
spot.formula(f)
```

```python
!ltlcross -t ltl2tgba -t "../Rab4/bin/ltl2dra  %f > %O" -t "../Rab4/bin/ltl2dgra  %f > %O" -f '{f}'
```

```python
spot.automaton('../Rab4/bin/ltl2dra "{}" |'.format(f))
```

The automaton does reject ${p_o}\emptyset^\omega$ which is from L(f). As the automaton produced by ltl2dgra seems to be correct, the bug is probably in the degeneralization.
